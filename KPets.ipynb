King County's Lost, Found, and Adoptable Pets: An Introduction

This is a short exploration of the open dataset provided by King County titled, "Lost, found, adoptable pets". I picked this data because I have a parrot, and the facebook groups I'm in frequently make posts about lost birds. This led me to wonder if there was an open dataset for lost animals. A simple google search led me to King County's open dataset repository, where I found the pet section (https://data.kingcounty.gov)

My goal with the data is to look at whether or not there is a difference in the proportions of dogs and cats that are considered "adoptable".

First, I wanted to gain an understanding of the type of data found within the dataset, so I looked at the beginning and end of the data

```python
import pandas as pd
lfa = pd.read_csv('LostFoundAdoptable.csv')

lfa.head() 
```

```python
lfa.tail()
```

Next, I wanted to obtain a greater undestanding of the total shape of the data:

```python
lfa.shape
```

This indicates that there are 402 rows and 25 columns. Upon physically reviewing the data, I found that there were different names for the same animal: Example: Cat vs. Barn Cat. Therefore, I needed to make sure I subsetted the data to contain the animal, and not a specific name or identifier. I used the following to subset all the data based on Animal Type, Gender, and Record Type (lost, found, or adoptable):

```python
PetData = (lfa[['animal_type', 'Animal_Gender', 'Record_Type']])
print(PetData)
```

I then performed the same function but with cats and dogs

```python
CatData = PetData[PetData['animal_type'].str.contains('Cat')]
print(CatData)
```

```python
DogData = PetData[PetData['animal_type'].str.contains('Dog')]
print(DogData)
```

I then wanted to find out how many dogs and cats were considered to be adoptable, so I subsetted these data further:

```python
AdoptableCat = CatData[CatData['Record_Type'] == 'ADOPTABLE']
print(AdoptableCat)
```

```python
AdoptableDog = DogData[DogData['Record_Type'] == 'ADOPTABLE']
print(AdoptableDog)
```

After reviewing the data, I developed the following question: at a alpha = 0.05 significance level, does sufficient evidence exist showing that the proportion of adoptable dogs is greater than that of adoptable cats? This creates the following hypotheses: 
Null hypothesis: There is no significant difference in the proportion of adoptability between dogs and cats
Alternate hypothesis: There is a significant difference in the proportion of adoptable dogs than adoptable cats.
Therefore, our analysis will be right tailed

```python
from statsmodels.stats.proportion import proportions_ztest
adopt = [len(AdoptableCat), len(AdoptableDog)]
n = [len(DogData), len(CatData)]
print(proportions_ztest(adopt, n))
```

The proportions_ztest indicates a z-score of 0.588, with a p-value of 0.57. Since the p-value 0.57 > 0.05, we can infer that there is no significant difference in the proportion of adoptability between cats and dogs.
